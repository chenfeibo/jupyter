```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.describe()
```

```python
iris.plot(kind="scatter", x="sepal_length", y="sepal_width")
```

```python
sns.pairplot(iris, hue='species', markers='+')
```

## Splitting data in train test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(iris[['sepal_length',	'sepal_width', 	'petal_length',	'petal_width']], iris[['species']], test_size=0.25)
x_train.head()
```

```python
print (x_train.shape)
print (x_test.shape)
print (y_train.shape)
print (y_test.shape)
```

## KNN classifier

```python
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression



knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train.values.ravel())
y_pred_knn = knn.predict(x_test)
metrics.accuracy_score(y_test, y_pred_knn)

```

## Logistic Regression Classifier

```python
logreg = LogisticRegression(solver='liblinear', max_iter=500)
logreg.fit(x_train, y_train.values.ravel())
y_pred_log_reg = logreg.predict(x_test)
print(metrics.accuracy_score(y_test, y_pred_log_reg))
```

```python
predicted = logreg.predict(iris[['sepal_length','sepal_width','petal_length', 'petal_width']])
predicted
```
